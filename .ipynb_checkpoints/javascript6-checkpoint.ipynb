# DOM - Encontrando elementos por su etiqueta (tag name)
> ## M茅todo document.getElementByTagName
> Este m茅todo permite encontrar todos los elementos HTML que correspondan a una misma etiqueta, se procede de la misma forma que en el ejemplo anterior, hay algunas consideraciones importantes:
>
> - La variable donde se almacenan las etiquetas encontradas es de tipo lista (array), aunque no se necesario declararla como tal
> - Los elementos encontrados ocupar谩n una posici贸n en la lista de acuerdo a su orden de aparici贸n en el documento HTML.
> - para acceder a un elemento en particular, se deben usar los 铆ndices, en JavaScript como en Python, la indexaci贸n inicia con el n煤mero cero.
>

>
> ## Ejemplo
> El siguiente c贸digo HTML presenta varias etiquetas de tipo _p(p谩rrafo)_, usaremos un script para encontrarlas y obtener el valor almacenado en alguna de ellas.

```python
%%html
<html>
<body>

<h2>Encontrando valores de elementos HTML por el nombre de la etiqueta</h2>

<p>Hola Mundo!</p>
<p>Ejemplo del uso del metodo <b>getElementsByTagName</b>.</p>

<p id="demo">Fin del ejemplo</p>


</body>
</html>
```

> A continuaci贸n, dentro del script el m茅todo _document.getElementByTagName_ se encargar谩 de realizar la b煤squeda y almacenar cada acierto en una posici贸n de la lista.

```python
%%html
<script>
var x = document.getElementsByTagName("p");
document.write('El texto en la primera etiqueta hallada (index 0) es: ' + x[0].innerHTML);
</script>
```

> ## Descripci贸n
>> Dentro de la variable x se creara la lista que ser谩 llenada por el metodo document.getElementsByTagName("p")<br>
>> Usando la sentencia document.write para escribir al final de todos los elementos el mensaje: "El texto en la primera etiqueta hallada (index 0) es: " <br>
>> Finalmente, se agrega con el operador "+" el primer elemento encontrado con dicha etiqueta usando la propiedad _innerHTML_, el cual corresponde a la posici贸n cero entre corchetes.
