## Generalidades de JavaScript
> Como lo asegura la Mozilla Foundation las páginas web donde nos ofrecen interacción ya sea accediendo a mapas, manipulando tablas, deslizadores, o simplemente llenando formularios tienen dentro de su código presente a JavaScript, si consideramos las paginas web como un pastel dividido en varias capas, JavaScript hace parte de la capa interior [Mozilla Fundation](https://developer.mozilla.org/es/docs/Learn/JavaScript/First_steps/Qu%C3%A9_es_JavaScript):

> - HTML es un lenguaje de marcado que usa la estructura para dar un sentido al contenido web, por ejemplo define párrafos, cabeceras, tablas, imágenes y vídeos en la página.
> - CSS es un lenguaje de reglas en cascada que usamos para aplicar un estilo a nuestro contenido en HTML, por ejemplo colocando colores de fondo, fuentes y marginando nuestro contenido en múltiples columnas.
> - JavaScript Es un lenguaje de programación que te permite crear contenido nuevo y dinámico, controlar archivos de multimedia, crear imágenes animadas y muchas otras cosas más. (Aunque, no todo, pero es increíble lo que puedes llegar a hacer con tan sólo unas pocas líneas de código de JavaScript).

## Qué hace JavaScript dentro de una página web?
> Tomado de [Mozilla Foundation](https://developer.mozilla.org/es/docs/Learn/JavaScript/First_steps/Qu%C3%A9_es_JavaScript): cuando abrimos una página web, el navegador (Chrome, Mozilla, Safari, Edge, Internet Explorer, etc) opera como una planta procesadora de materia prima (las lineas de código) las interpreta y las convierte en textos, elementos gráficos, colores e interacciones. Estas últimas son las que están realizadas en JavaScript, después de presentar todos los códigos nos lo muestra como una página web. 
>Cuando insertamos codigo de JavaScript en nuestra página, éste es reconocido en orden, es decir, primero se ejecutan las lineas de arriba y cada una de ellas de izquierda a derecha, por esto se debe ser cuidado en plantear un orden lógico en nuestro código.
> ### Código del lado del usuario
> El termino ejecución del lado de usuario (o del cliente) significa que el código se ejecuta en el computador del usuario,  este proceso lo realiza el navegador desde el cual estamos abriendo la página, el cual sirve de interprete de tal código. JavaScript es un lenguaje interpretado que se ejecuta del lado del cliente.

## Insertando código de JavaScript en nuestra página web
Existen básicamente dos formas de poder agregar código de JavaScript en nuestro sitio web:
> ## 1. En el mismo archivo HTML (JavaScript por dentro o embebido)
Se llama código embebido y se escribe dentro de las etiquetas del _body_ (cuerpo de la página): 

```python
%%html
<script>
// acá va nuestro código JavaScript 
</script>
```

>> En este caso se deben tener algunas reglas:
>> - Debe ir después de las etiqueta _body_
>> - Debe ubicarse después de los elementos HTML que deseamos modificar
>> - Debe cerrarse antes de la etiqueta _body_ de cierre

> ## 2. En un archivo externo (JavaScript externo o por fuera)
Se deben cumplir los siguientes requisitos:
>> 1. El archivo con nuestro código JavaScript debe estar guardado preferiblemente en la misma ubicación que nuestro archivo HTML
>> 2. El numbre del archivo con el código de JavaScript debe tener la extensión __.js__, por ejemplo __MiCodigoJava.js__
>> 3. En la parte del documento HTML donde deseamos insertar el código de JavaScript debemos escribir:

```python
%%html
<script src="nombre_archivo.js"></script>
```

## Ejemplo
En el siguiente ejemplo, se escribirá un código JavaScript embebido (por dentro), que al desplegar la pagina muestre un cuadro de dialogo preguntándonos nuestro nombre, al ingresar la informacion se mostrará un mensaje saludando al usuario usando tres métodos diferentes
> ## Construyendo el documento HTML
El primer paso, consiste en elaborar nuestro documento contenedor del código JavaScript, para ellos usaremos el bloc de notas escribiendo las etiquetas tradicionales de html:

```python
%%html
<html>
<head>
    <title> Primer Ejemplo de JavaScript </title>
</head>
```

> En seguida, aplicaremos un color de fondo usando la codificación hexadecimal (debe buscar un color diferente al usado en el ejemplo) y aplicaremos un título de cabecera _h1_

```python
%%html
<body bgcolor="#FF5733">
<h1> Botón Interactivo </h1>
```

> En un color de texto diferente y en cabecera _h2_ escribiremos el siguiente texto (use un color distinto al del ejemplo):

```python
%%html
<font color="#33B7FF">
    <h2 id="texto"> Mensaje de bienvenida</h2>
</font>
```

> ## Escribiendo el Código de JavaScript
> Nótese que se ha agregado después de _h2_ el argumento "id", como su nombre lo indica sirve para asignarle un identificador a la etiqueta HTML, el cual es necesario si queremos modificarla con un código de JavaScript.
Procedemos a agregar nuestro script usando las siguientes sentencias:
> - prompt: para mostrar un cuadro de entrada de datos, en este caso será el nombre del usuario es decir una cadena (string)
>> ### Primer método
>> - innerHTML: este metodo permite cambiar (rescribir) el contenido de un elemento HTML

```python
%%html
<script>
    var nombre = prompt("ingresa tu nombre");
    document.getElementById("texto").innerHTML = "hola! " + nombre;
</script>
```

> Si queremos en lugar de cambiar el contenido de un elemento HTML aparezca una ventana con el mensaje de saludo, debemos modificar (o agregar) a la linea 4 del código anterior la sentencia:
>> ### Segundo Método
>> - alert: sirve para mostrar una ventana emergente con un mensaje

```python
%%html
    alert("Hola! " + nombre);
```

> Por último, si queremos escribir algo dentro del documento sin modificar un elemento HTML, podemos hacer uso de la sentencia:
>> ### Tercer Método
>> - _document.write_, sirve para escribir direcamente sobre el documento HTML, por ejemplo para que el mensaje aparezca debajo del elemento _h2_, agregamos:

```python
%%html
    document.write("Hola! " + nombre);
```

> __Aclaración__: Los diferentes métodos expuestos hasta el momento no son excluyentes y pueden ser usados en forma combinada dependiendo del contexto.
